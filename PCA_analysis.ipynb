```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

def generate_data(n_samples=100, n_features=3, random_state=42):
    np.random.seed(random_state)
    mean = np.zeros(n_features)
    cov = np.array([[1, 0.8, 0.6], [0.8, 1, 0.5], [0.6, 0.5, 1]])  # Correlated features
    data = np.random.multivariate_normal(mean, cov, size=n_samples)
    return data

def perform_pca(data, n_components=2):
    scaler = StandardScaler()
    data_scaled = scaler.fit_transform(data)

    pca = PCA(n_components=n_components)
    principal_components = pca.fit_transform(data_scaled)

    return pca, principal_components

def plot_pca_results(principal_components):
    plt.figure(figsize=(8, 6))
    plt.scatter(principal_components[:, 0], principal_components[:, 1], alpha=0.7, edgecolors='k')
    plt.xlabel('Principal Component 1')
    plt.ylabel('Principal Component 2')
    plt.title('PCA Projection')
    plt.grid(True)
    plt.show()


data = generate_data(n_samples=200, n_features=3)
pca, principal_components = perform_pca(data, n_components=2)

print("Explained Variance Ratios:", pca.explained_variance_ratio_)
plot_pca_results(principal_components)
```
